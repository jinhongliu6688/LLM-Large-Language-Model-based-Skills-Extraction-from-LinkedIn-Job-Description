# Milestone 2 Evaluation

## Data

```python
import pandas as pd
```

```python
gold_df = pd.read_csv('final_annotated_data.csv')
gold_df = gold_df.reset_index()
gold_df
```

```python
sys_df = pd.read_csv('GPT3_data_for_evaluation.csv')
sys_df = sys_df.reset_index()
sys_df
```

```python
gold_df.job_description.tolist() == sys_df.job.to_list()
```

```python
gold_dict = dict(zip(gold_df.index, gold_df.skills))
```

```python
for key, value in gold_dict.items():
    gold_dict[key] = gold_dict[key].split(",")
```

```python
sys_dict = dict(zip(sys_df.index, sys_df.skills))
```

```python
for key, value in sys_dict.items():
    try:
        sys_dict[key] = sys_dict[key].replace("'", '').replace("[", "").replace("]", "").replace(" ", "").split(",")
    except:
        sys_dict[key] = []
```

```python
sys_dict.keys() == gold_dict.keys()
```

## Evaluation Result

```python
tp = 0
fp = 0
fn = 0

for key, value in sys_dict.items():
    gold_set = set(gold_dict[key])
    sys_set = set(x.lower() for x in sys_dict[key])
    
    # print(gold_set)
    # print(sys_set)
    # print()

    tp += len(sys_set & gold_set)
    fp += len(sys_set - gold_set)
    fn += len(gold_set - sys_set)
            
precision = tp/(tp+fp) * 100
recall = tp/(tp+fn) * 100

print(f"Precision: {precision}%")
print(f"Recall: {recall}%")
print(f"F1 score: {2*precision*recall/(precision+recall)}%")
```
